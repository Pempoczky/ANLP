```python
from transformers import pipeline

generator_bias = pipeline('text-generation', model='Qwen/Qwen2.5-0.5B')
generator_debias = pipeline('text-generation', model='LeoZotos/Qwen2.5-0.5B_debiased')
```

```python
sentences_en = ['There were some unexpected and unforeseen', 'I still remember the day that', 'Do you know where i left my car keys? I swear I left them', 'What was the name of the doctor again? I think it was']
sentences_nl = ['Er waren wat onverwachte en onvoorziene', 'Ik herriner nog steeds de dag dat', 'Weet je waar ik mijn autosleutels heb gelaten? I zweer dat ik ze', 'Hoe heette de dokter ook alweer? I dacht dat het']
max_tokens = 50

for prompt in sentences_en:
  result = generator_bias(prompt, max_new_tokens=max_tokens)
  print(result[0]['generated_text'])

print("\n-------\n")

for prompt in sentences_nl:
  result = generator_bias(prompt, max_new_tokens=max_tokens)
  print(result[0]['generated_text'])

print("\n-------\n")

for prompt in sentences_en:
  result = generator_debias(prompt, max_new_tokens=max_tokens)
  print(result[0]['generated_text'])

print("\n-------\n")

for prompt in sentences_nl:
  result = generator_debias(prompt, max_new_tokens=max_tokens)
  print(result[0]['generated_text'])
```
